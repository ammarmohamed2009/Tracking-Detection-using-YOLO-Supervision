```python
from ultralytics import YOLO
import cv2
```

```python
model = YOLO('yolov8m.pt')
```

```python
v_path = 'Desktop/AI_803/medo.mp4'
cap = cv2.VideoCapture(v_path)
```

***Detection***

```python
while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)

    annotated_frames = results[0].plot()

    cv2.imshow("Detection", annotated_frames)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

***Tracking***

```python
model = YOLO('yolov8m.pt')
v_path = 'Desktop/AI_803/medo.mp4'
cap = cv2.VideoCapture(v_path)
```

```python
import supervision as sv
tracker = sv.ByteTrack()
box_annotator = sv.BoxAnnotator()
```

```python
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

fourcc = cv2.VideoWriter_fourcc(*'mp4v') 
out = cv2.VideoWriter('output_tracking_video.mp4', fourcc, fps, (frame_width, frame_height))

while True:
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)[0]
    detections = sv.Detections.from_ultralytics(results)
    tracked_detections = tracker.update_with_detections(detections)

    # Drawing the boxes
    frame = box_annotator.annotate(scene=frame, detections=tracked_detections)

    for box in range(len(tracked_detections)):
        x1, y1, x2, y2 = tracked_detections.xyxy[box].astype(int)
        class_id = tracked_detections.class_id[box]
        confidence = tracked_detections.confidence[box]
        track_id = tracked_detections.tracker_id[box]

        # color logic
        my_color = (0, 255, 0) if class_id == 0 else (0, 0, 255)
        
        label = f"{model.names[class_id]} {confidence:.2f} ID:{track_id}"
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, my_color, 2)

    # SAVE THE FRAME
    out.write(frame) 

    cv2.imshow("Tracking", frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
out.release()
cv2.destroyAllWindows()
```
